```python
import pandas as pd

# load in data
file_path = 'AppleStore.csv'

df = pd.read_csv(file_path)

file_path2 = 'appleStore_description.csv'

df2 = pd.read_csv(file_path2)

df.head(5)
```

```python
# merge 2 dfs
appdf = pd.merge(df, df2, on='id')
```

```python
# drop columns not being used
columns_to_drop = ['currency', 'rating_count_tot', 'user_rating', 'user_rating_ver', 'rating_count_ver', 'sup_devices.num', 'ipadSc_urls.num', 'lang.num', 'lang.num', 'vpp_lic', 'size_bytes_y', 'track_name_y']
appdf = appdf.drop(columns=columns_to_drop)
```

```python
# rename them
appdf.rename(columns={'track_name_x': 'name'}, inplace=True)
appdf.rename(columns={'size_bytes_x': 'size_bytes'}, inplace=True)
appdf.rename(columns={'size_bytes_x': 'size_bytes'}, inplace=True)
appdf.rename(columns={'prime_genre': 'genre'}, inplace=True)
appdf.rename(columns={'app_desc': 'description'}, inplace=True)

appdf.head(5)
```

```python
# clean up the df
appdf.drop(columns=['Unnamed: 0'], inplace=True)
appdf.drop(columns=['id'], inplace=True)
appdf.rename(columns={'limited_genre': 'genre'}, inplace=True)
appdf.head(5)
```

```python
appdf.head(20)
```

```python
appdf['genre'].unique()
```

```python

# find out how many are in each genre
genre_counts = appdf['genre'].value_counts()

# take the top 5 genres
top_genres = genre_counts.head(5).index.tolist()

# if its not in the top 5, set it to other
def map_to_other(genre):
    return genre if genre in top_genres else 'Other'

# apply the map to other function 
appdf['genre'] = appdf['genre'].apply(map_to_other)


appdf.head(50)
```

```python
# limit to the top 100 rows
appdf_top_100 = appdf.head(100)

# export to JSON file
appdf_top_100.to_json('apps.json', orient='records', lines=True)
```
